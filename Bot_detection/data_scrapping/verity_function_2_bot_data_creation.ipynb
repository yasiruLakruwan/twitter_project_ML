```python
import pandas as pd
from faker import Faker
import random
import spacy
```

```python
# Initialize Faker and spaCy
fake = Faker()
nlp = spacy.load("en_core_web_sm")
```

```python
# Function to generate a random tweet
def generate_tweet():
    return fake.sentence(nb_words=random.randint(5, 20))

# Function to count mentions and hashtags
def count_mentions_hashtags(text):
    mentions = text.count('@')
    hashtags = text.count('#')
    return mentions, hashtags
```

```python
# Function to calculate ORG and PERSON percentages
def calculate_entity_percentage(text):
    doc = nlp(text)
    total_tokens = len(doc)
    org_count = sum(1 for ent in doc.ents if ent.label_ == 'ORG')
    person_count = sum(1 for ent in doc.ents if ent.label_ == 'PERSON')

    if total_tokens > 0:
        org_percentage = (org_count / total_tokens) * 100
        person_percentage = (person_count / total_tokens) * 100
    else:
        org_percentage = 0
        person_percentage = 0

    return org_percentage, person_percentage

# Generate 100,000 synthetic tweets
num_tweets = 100000
data = []
```

```python
for _ in range(num_tweets):
    tweet = generate_tweet()
    mentions, hashtags = count_mentions_hashtags(tweet)
    org_percentage, person_percentage = calculate_entity_percentage(tweet)
    # Label the tweet as bot (1)
    bot_label = 1

    data.append({
        'tweet': tweet,
        'mentions': mentions,
        'hashtags': hashtags,
        'ORG_percentage': org_percentage,
        'PERSON_percentage': person_percentage,
        'BotScoreBinary': bot_label
    })
```

```python
# Create DataFrame
df = pd.DataFrame(data)
```
