```python
from ntscraper import Nitter
```

```python
scraper = Nitter(log_level = 1,skip_instance_check=False)
```

```python
tweets1 = scraper.get_tweets("ceylon_bot_SRI",mode="user",number=100)
```

```python
from pprint import pprint
```

```python
pprint(tweets1)
```

```python
tweets1.keys()
```

```python
# Flatten the tweet data
def flatten_tweet(tweet):
    return {
        'date': tweet.get('date', ''),
        'text': tweet.get('text', ''),
        'link': tweet.get('link', ''),
        'likes': tweet.get('stats', {}).get('likes', 0),
        'retweets': tweet.get('stats', {}).get('retweets', 0),
        'comments': tweet.get('stats', {}).get('comments', 0),
        'user_name': tweet.get('user', {}).get('name', ''),
        'user_username': tweet.get('user', {}).get('username', ''),
        'user_avatar': tweet.get('user', {}).get('avatar', ''),
    }

# Check if 'tweets' key exists in tweets1
if 'tweets' in tweets1:
    # Flatten all tweets
    flattened_tweets = [flatten_tweet(tweet) for tweet in tweets1['tweets']]

    # Convert to DataFrame
    df = pd.DataFrame(flattened_tweets)

    # Save to CSV
    df.to_csv('tweets1.csv', index=False)

    print("CSV file 'tweets1.csv' has been created.")
else:
    print("No tweets found or unexpected data structure.")
```

```python
import pandas as pd
```

```python
df1 = pd.read_csv("/content/drive/MyDrive/tweets1.csv")
df2 = pd.read_csv("/content/drive/MyDrive/tweets2.csv")
```
